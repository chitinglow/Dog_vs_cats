```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
sns.set_style('darkgrid')
plt.rcParams["figure.figsize"] = (30,30)
```

```python
data = pd.read_csv("cats_vs_dogs.csv", index_col=0)
data.describe()
```

```python
data.info()
```

```python
dogs = data[['state','percent_dog_owners', 'dog_population','avg_dogs_per_household', 'n_dog_households']]
cats = data[['state','percent_cat_owners', 'cat_population', 'avg_cats_per_household', 'n_cat_households']]
dogs.head()
```

```python
sort = dogs.sort_values(by = ['percent_dog_owners'],ascending=False)
sort = sort[:10]
sort
```

```python
f, ax = plt.subplots(figsize = (10,10))
_ = sns.barplot(x = 'state', y = 'percent_dog_owners', color = 'red', data = sort)
plt.xlabel('State', size = 12)
plt.ylabel("% of Dog Owners", size =12)
plt.xticks(rotation = 30, size = 13)
plt.yticks(size = 13)
```

```python
sort = cats.sort_values(by = ['percent_cat_owners'],ascending=False)
sort = sort[:10]
sort
f, ax = plt.subplots(figsize = (10,10))
_ = sns.barplot(x = 'state', y = 'percent_cat_owners', color = 'red', data = sort)
plt.xlabel('State', size = 12)
plt.ylabel("% of Cat Owners", size =12)
plt.xticks(rotation = 30, size = 13)
plt.yticks(size = 13)
```

```python
all_all = dogs.sort_values(by = ['avg_dogs_per_household'],ascending=False)
_ = sns.barplot(x = 'state', y = 'avg_dogs_per_household', color = 'red', data = all_all)
plt.xticks(rotation = 90, size = 10)
```

```python
all_all = cats.sort_values(by = ['avg_cats_per_household'],ascending=False)
_ = sns.barplot(x = 'state', y = 'avg_cats_per_household', color = 'red', data = all_all)
plt.xticks(rotation = 90, size = 10)
```

```python
data['diff'] = data.percent_dog_owners - data.percent_cat_owners
data = data.sort_values(by = ['diff'])
data['positive'] = data['diff'] > 0
data.head()
```

```python
red_patch = mpatches.Patch(color='red', label='Dog States')
blue_patch = mpatches.Patch(color='blue', label='Cat States')

data.plot.barh(x = 'state', y = 'diff', color=data.positive.map({True: 'r', False: 'b'}))
plt.title("Dog Versus Cat Orientation By State", size = 30)
plt.xlabel("Percent Difference", size = 30)
plt.ylabel("States", size = 30)
plt.xticks(size = 30)
plt.yticks(size = 28)
plt.xlim([-15,20])
plt.legend(handles=[red_patch, blue_patch], prop={'size': 25})
ax.get_figure()
plt.savefig('Dog_vs_Cat.jpg')
```
